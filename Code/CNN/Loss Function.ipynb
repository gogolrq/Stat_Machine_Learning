```python
import torch
from torch import nn
import math
torch.manual_seed(1)

m = nn.Sigmoid()
sfm = nn.Softmax(dim=1)
loss = nn.BCELoss()
loss2 = nn.BCEWithLogitsLoss()
input = torch.randn(3, 2, requires_grad=True) 
target = torch.empty(3).random_(2) #target as lable
pinput = sfm(input) # change input to probability vector

print(input)
print(pinput)

print(pinput[:,1])
print(target)

output = loss(pinput[:,1], target)
#-log(1-0.4027)-log(0.6364)-log(0.5710)
print((-math.log(1-0.4027)-math.log(0.6364)-math.log(0.5710))/3)
print(output)

```

```python
torch.manual_seed(1)

m = nn.Sigmoid()
input = torch.randn(3, requires_grad=True)
target = torch.empty(3).random_(2)
pinput = m(input)
print(input)
print(pinput)
print(target)

output = loss(pinput, target)
print(output)


output2 = loss2(input, target)

print(output2)

# -log(0.6596)-log(1-0.5663)-log(1-0.5154)
print((-math.log(0.6596)-math.log(1-0.5663)-math.log(1-0.5154))/3)

```

```python
torch.manual_seed(1)
# Example of target with class indices
loss = nn.CrossEntropyLoss()
input = torch.randn(3, 5, requires_grad=True)
pinput = sfm(input)
target = torch.empty(3, dtype=torch.long).random_(5)
output = loss(input, target)

print(input)
print(pinput) 
print(target)
print(output)

# -log(0.1563)-log(0.1035)-log(0.2015) 
print((-math.log(0.1563)-math.log(0.1035)-math.log(0.2015))/3)
```

```python
torch.manual_seed(1)
# Example of target with class probabilities
input = torch.randn(3, 5, requires_grad=True)
pinput = sfm(input)
target = torch.randn(3, 5).softmax(dim=1)
output = loss(input, target)

print(input)
print(target)
print(output)
```
