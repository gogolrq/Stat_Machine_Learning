```python
img = Image.open("gengar.jpg")
#plt.subplot(1,3,1)
plt.imshow(img)

img_pixels = np.asarray(img) #img_pixels

print(img_pixels.shape) # numpy [width, height, channels], jpg has 3 channels
```

```python
# Gray=0.2125*R+0.7154*G+0.0721*B
arr_gray = 0.2125*img_pixels[:,:,0]+0.7154*img_pixels[:,:,1]+0.0721*img_pixels[:,:,2]
print(arr_gray.shape)

img_gray = Image.fromarray(arr_gray)

plt.axis("off")
plt.imshow(img_gray)
```

```python
convert_tensor = transforms.ToTensor()
img_tensor = convert_tensor(img)
print(img_tensor.shape) # tensor [channels, width, height]

plt.axis("off")
plt.imshow(img_tensor.permute(1, 2, 0))
```
