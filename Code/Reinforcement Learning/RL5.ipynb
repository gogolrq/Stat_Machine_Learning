```python
import gymnasium as gym
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
from collections import defaultdict

env = gym.make("Blackjack-v1", natural=False, sab=False)
print("Action space:", env.action_space)
print("Example reset:", env.reset())
```

```python
# collect data using a uniform policy
def collect_blackjack_data(env, n_episodes=50000):
    rows = []

    for ep in range(n_episodes):
        state, info = env.reset()
        done = False

        while not done:
            action = env.action_space.sample()   # uniform over {0,1}
            next_state, reward, terminated, truncated, info = env.step(action)
            done = terminated or truncated

            rows.append({
                "player_sum": state[0],
                "dealer_showing": state[1],
                "usable_ace": int(state[2]),
                "action": action,
                "reward": reward,
                "next_player_sum": next_state[0],
                "next_dealer_showing": next_state[1],
                "next_usable_ace": int(next_state[2]),
                "done": int(done),
            })

            state = next_state

    df = pd.DataFrame(rows)
    return df
```

```python
# run data collection and save to file
df = collect_blackjack_data(env, n_episodes=50000)

print("Number of transitions:", len(df))
display(df.head())

df.to_csv("blackjack_uniform_data.csv", index=False)
print("Saved to blackjack_uniform_data.csv")
```
